```python
import numpy as np # linear algebra
import pandas as pd 
```

```python
ds = pd.read_csv('dataset_q2_q4.csv')
```

```python
X = ds[['x1', 'x2', 'x3']]
y = ds['y_label']

X.shape, y.shape
```

```python
def find_theta(X, y):
    
    m = X.shape[0] # Number of training examples. 
    X = np.append(X, np.ones((m,1)), axis=1)    

    X_t = X.T
    #theta = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(y,X))
    z = np.ones((np.shape(X)[0],1))
    x_new = np.hstack((z,X))
    #theta=np.linalg.inv(x_new.T.dot(x_new)).dot(x_new.T).dot(y)

    # The Normal Equation
    theta = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    
    return theta
```

```python
def predict(X):
    
    # Appending a cloumn of ones in X to add the bias term.
    X = np.append(X, np.ones((X.shape[0],1)), axis=1)
    
    # preds is y_hat which is the dot product of X and theta.
    preds = np.dot(X, theta)
    
    return preds
```
